```python
mask = np.array([1,1,1,0], dtype = bool)
print(mask)
print(mask.shape)
row = np.array([0, 2])
print(row[:, np.newaxis].shape)
print(x[row[:, np.newaxis], mask])
```

генерация на плоскости точек

```python
rng = np.random.default_rng(seed=1)
```

```python
x = rng.multivariate_normal([0, 0], [[1, 2], [2, 5]], 100)
print(x.shape)
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```python
np.random.seed(0)
inx = np.random.choice(100, 30, replace=False)
print(inx)
```

```python
select = x[inx]
plt.scatter(x[:, 0], x[:, 1], alpha=0.3)
plt.scatter(select[:, 0], select[:, 1], s = 200, facecolor = "none", edgecolor = 'black')
```

```python
x = np.arange(10)
print(x)
inx = np.array([2, 8, 4, 1])

x[inx] = 90
print(x)
```

```python
x = np.arange(10)
print(x)
inx = np.array([4, 4, 4, 4, 4, 1])

x[inx] += 1
print(x)
```

```python
x = np.arange(10)
print(x)

np.add.at(x, inx, 1)
print(x)
```

разбиение данных на интервалы

```python
rng = np.random.default_rng(seed=1)
x = rng.integers(100, size=100)
print(x[:10])
```

```python
bins = np.linspace(0, 100, 11)
print(bins)
```

```python
counts = np.zeros(11)
print(counts)
```

```python
i = np.searchsorted(bins, x)
print(i[:10])
```

```python
np.add.at(counts, i, 1)
print(counts)
```

```python
print(sum(counts))
```

сортировка массивов

```python
a = [3,2,5,1,4,6,78,64,23,5,3,1435,6]
print(sorted(a))
print(a)
```

```python
a.sort()
print(a)
```

```python
a = [3,2,5,1,4,6,78,64,23,5,3,1435,6]
x = np.array(a)
print(x)
```

```python
print(np.sort(x))
print(x)

x.sort()
print(x)
```

```python
x = np.array(a)
inx = np.argsort(x)
print(x)
print(np.sort(x))
print(inx)
print(x[inx])
```

```python
rng = np.random.default_rng(seed=1)
x = rng.integers(0, 10, size=(4, 6))
print(x)

print(np.sort(x, axis=1))
print(np.sort(x, axis=0))
```

```python
name = ['Ирина', 'Виталий', 'Олег', 'Саша']
age = [25, 17, 52, 44]
weight = [55.0, 57, 78, 72]

i = 1
print(name[i], age[i], weight[i])
```

```python
data = np.zeros(
    4,
    dtype = {
        "names": ("name_", "age_", "weight_"),
        "formats": ("U10", "i4", "f8")
    })

print(data.dtype)
```

```python
data["name_"] = name
data["age_"] = age
data["weight_"] = weight

print(data)
```

```python
print(data["name_"])
```

```python
print(data[0])
```

```python
print(data[-1]["name_"])
```

```python
data_rec = data.view(np.recarray)
print(data_rec)
```

```python
print(data_rec.name_)
print(data_rec[-1].name_)
```

```python
print(data["age_"] < 30)
```

```python
print(data[data["age_"] < 30])
```

```python
print(data[data["age_"] < 30]["name_"])
```
