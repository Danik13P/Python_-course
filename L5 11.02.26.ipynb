Атрибуты - индексаторы

```python
dict1 = {
    'A':10,
    'B':20,
    'C':30,
    'D':40,
    'E':50,
    'F':60
}
data_dict1 = pd.Series(dict1)
print(data_dict1)

dict2 = {
    'A':10,
    'B':20,
    'C':30,
    'D':40,
    'E':50,
    'H':75
}
data_dict2 = pd.Series(dict2)
print(data_dict2)

df = pd.DataFrame({"dict_01": data_dict1, "dict_02": data_dict2})
print(df)

print(df["dict_01"])
print(df.dict_01)

df["new"] = df["dict_01"]
print(df)

df["new"] = df["dict_01"] / df["dict_02"]
print(df)

print(df.values)

print(df["dict_01"]) # столбец
print(df.values[0])  # строка

print(df)

print(df.loc["A":"C", :"dict_02"])
print(df.iloc[:3, :2])

print(df.loc[df.dict_02 > 30, ["new1", "dict_01"]])

df.loc[df.dict_02 > 30, ["new1", "dict_01"]] = 36
print(df)
```

Транслирование

```python
A = rng.integers(10, size=(3,4))
print(A[0])

print(A - A[0])

df = pd.DataFrame(A, columns=["A","B","C","D"])
print(df)

print(df - df.iloc[0])

print(df.substract(df["A"], axis=0))
```
